3.1.2 1변량 데이터의 관리

```python
fish_data = np.array([2, 3, 3, 4, 4, 4, 4, 5, 5, 6])
fish_data
```

3.1.3 합계와 샘플사이즈

```python
# scipy.sum well be removed
print(sp.sum(fish_data))
```

```python
print(np.sum(fish_data))
```

```python
len(fish_data)
```

3.1.4 평균값(기댓값)

```python
N = len(fish_data)
sum_value = np.sum(fish_data)
mu = sum_value / N
mu
```

```python
print(sp.mean(fish_data))
```

```python
print(np.mean(fish_data))
```

3.1.5 표본분산

```python
sigma_2_sample = np.sum((fish_data - mu) ** 2) / N
sigma_2_sample
```

```python
print(f"fish_data             : {fish_data}")
print(f"fish_data - mu        : {fish_data - mu}")
print(f"(fish_data - mu) ** 2 : {(fish_data - mu) ** 2}")
```

```python
np.sum((fish_data - mu) ** 2)
```

```python
np.var(fish_data, ddof=0)
```

3.1.6 불편분산

```python
# 왜 표본 평균을 활용한 분산 계산은 분산을 과소추정하는 경향이 있는가?
sigma_2 = np.sum((fish_data - mu) ** 2) / (N - 1)
sigma_2
```

```python
np.var(fish_data, ddof=1)
```

3.1.7 표준편차

```python
sigma = np.sqrt(sigma_2)
sigma
```

```python
np.std(fish_data, ddof=1)
```

3.1.8 표준화

```python
fish_data - mu
```

```python
np.mean(fish_data - mu)
```

```python
fish_data / sigma
```

```python
np.std(fish_data / sigma, ddof=1)
```

```python
standard = (fish_data - mu) / sigma
```
