6.1.6 이항분포(실습)

```python
print(sp.stats.binom.pmf(k=1, n=2, p=0.5))
```

```python
np.random.seed(1)
print(sp.stats.binom.rvs(n=10, p=0.2, size=5))
```

```python
binomial = sp.stats.binom(n=10, p=0.2)

np.random.seed(1)
rvs_binomial = binomial.rvs(size=10000)

m = np.arange(0, 10, 1)
pmf_binomial = binomial.pmf(k=m)

# distplot > histplot
sns.histplot(rvs_binomial, bins=m, stat="density", kde=False)
plt.plot(m, pmf_binomial)
```

```python
print(sp.stats.poisson.pmf(k=2, mu=5))
```

```python
np.random.seed(1)
print(sp.stats.poisson.rvs(mu=2, size=5))
```

```python
poisson = sp.stats.poisson(mu=2)

np.random.seed(1)
rvs_poisson = poisson.rvs(size=10000)
pmf_poisson = poisson.pmf(k=m)
sns.histplot(rvs_poisson, bins=m, stat="density", kde=False)
plt.plot(m, pmf_poisson)
```

```python
N = 100000000
p = 0.00000002

binomial_2 = sp.stats.binom(n=N, p=p)
pmf_binomial_2 = binomial_2.pmf(k=m)

plt.plot(m, pmf_poisson, color='red')
plt.plot(m, pmf_binomial_2, color='blue', linestyle='dashed')
```
